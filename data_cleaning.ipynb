```python
def clean_country(raw):
    ignore = ['Afr', 'Asia', 'ICC', 'XI']
    parts = raw.split('/')
    
    # Filter out generic tokens
    parts = [p.strip().upper() for p in parts if p.strip().upper() not in ignore]
    
    # Handle specific combinations
    if raw == 'ENG/IRE':
        return 'IRE'
    elif raw == 'AUS/SA':
        return 'SA'
    
    # If only one valid country remains
    if len(parts) == 1:
        return parts[0]
    elif len(parts) >= 2:
        return parts[-1]  # default: keep the last valid country
    else:
        return None
```

```python
data['country']=data['country'].apply(clean_country)
```

## Creating of Model

```python
from sklearn.preprocessing import MinMaxScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.metrics.pairwise import cosine_similarity
```

```python
numerical_cols = ['matches', 'innings', 'runs', 'highest score', 'average',
                  'strike rate', 'duration', 'not out ratio', 'duck rate',
                  'century frequency', 'century conversion rate', 'balls faced per innings']
categorical_cols = ['country']

# Build MinMaxScaler pipeline
minmax_preprocessor = ColumnTransformer([
    ('num', MinMaxScaler(), numerical_cols),
    ('cat', OneHotEncoder(), categorical_cols)
])
```

```python
pipeline = Pipeline([
    ('preprocessor', minmax_preprocessor)
])
```

```python
transformed_data = pipeline.fit_transform(data)
```

```python
pipeline.fit(data)
```

```python
elite_transformed = pipeline.transform(data)
```

```python
new_player_data = {
    'matches': [350],
    'innings': [297],
    'runs': [10773],
    'highest score': [183],
    'average': [50.57],
    'strike rate': [87.56],
    'country': ['India'],
    'duration': [15],
    'not out ratio': [0.28],
    'duck rate': [0.03],
    'century frequency': [0.03],
    'century conversion rate': [0.12],
    'balls faced per innings': [41.42]
}
new_player = pd.DataFrame(new_player_data)
new_player_transformed = pipeline.transform(new_player)
```

```python
similarities = cosine_similarity(new_player_transformed, elite_transformed)[0]
average_similarity = similarities.mean()
scaled_score = round(average_similarity * 100, 2)
```

```python
print("Cosine Similarity Talent Score (0â€“100):", scaled_score)
```

```python
def classify_talent_score(score):
    """
    Classify player based on cosine similarity talent score (0â€“70 scale).
    """
    if score >= 60:
        return 'Future Top Player ðŸŒŸ'
    elif score >= 50:
        return 'Promising Talent ðŸŸ¡'
    elif score >= 30:
        return 'Needs Development âšª'
    else:
        return 'Not Close Yet ðŸ”´'
```

```python
print(classify_talent_score(scaled_score))
```
